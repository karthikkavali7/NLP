```python
import nltk
import spacy
```

```python
text = "hello iam karthik from sr university warangal iam pursuing second year cse aiml"
```

```python

lower_text = text.lower()
print(f"Lowercase text: {lower_text}")

word_count = len(lower_text.split())
print(f"Number of words: {word_count}")

nlp = spacy.load("en_core_web_sm")
doc = nlp(lower_text)
lemmatized_words = [token.lemma_ for token in doc]
lemmatized_text = " ".join(lemmatized_words)
print(f"Lemmatized text: {lemmatized_text}")
```
